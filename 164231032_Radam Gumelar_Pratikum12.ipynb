2. Cobalah kode berikut untuk menemukan jalur dari node 1 ke node 5. Jelaskan hasil
keluarannya<br>


```python
def find_all_path(graph, node_awal, node_akhir, path=[]):
    path = path + [node_awal]
    if node_awal == node_akhir:
        print("===== Jalur yang ditemukan:", path)
        return [path]
    if node_awal not in graph:
        return []
    paths = []

    for node in graph[node_awal]:
        print("Path saat ini", path)
        print("Node tujuan berikutnya adalah %s dari node sebelumnya yaitu %s" %(node, node_awal))

        if node not in path:
            newpaths = find_all_path(graph, node, node_akhir, path)
            for newpath in newpaths:
                paths.append(newpath)
    return paths

find_all_path(graph, "1","5")
```

**Penjelasan Program**<br>
program diatas adalah fungsi untuk mencari/menemukan semua jalur path yang mungkin dari `node_awal` sampai ke `node_akhir`<br>
fungsi `find_all_path` memiliki 4 parapeter:
- `graph` = ini untuk menyimpan struktur graph
- `node_awal` = untuk menyimpan nilai node awal yg ingin dicari
- `node_akhir` = untuk menyimpan nilai node akhir yang ingin ditemukan jalurnya
- `path` = ini isinya list kosong yg digunakan untuk menyimpan jalur saat kita melakukan pencarian pada program nantinya

penjelasan isi fungsi:
- `path = path + [node_awal]` = menambahkan nilai node awal ke dalam path
- `if node_awal == node_akhir:` = jika nilai node_awal sama dengan nilai node_akhir maka akan menampilkan output `"===== Jalur yang ditemukan:", path)`, kemudian fungsi mengembalikan nilai list pada `path`
- `if node_awal not in graph:` = jika node_awal **tidak** ditemukan pada graph maka akan mengembalikan nilai list kosong. ini maksudnya node_awal yg ingin dicari tidak ada pada graph
- `paths = []` = ini untuk menginisialisasi list kosong. untuk menyimpan semua jalur yang akan ditemukan
- `for node in graph[node_awal]:` = looping untuk nilai pada graph[indexnya yang diisi dengan node_awal] akan menampilkan output
  - "Path saat ini", path
  - "Node tujuan berikutnya adalah [node] dari node sebelumnya yaitu [node_awal]"
- `if node not in path:` = jika tetangga yang dikunjungi tidak ada di list path maka akan memanggil fungsi `find_all path` lagi
  - `newpaths = find_all_path(graph, node, node_akhir, path)` = variabel untuk menyimpan fungsi yg akan dipanggil lagi
  - `for newpath in newpaths` = looping untuk nilai newpath
    - `paths.append(newpath)` = menambahkan jalur yang ditemukan ke dalam list
- `return paths` = mengembalikan semua nilai pada paths (ini akan menampilkan semua kemungkinan jalur yang bisa ditempuh dari node_awal ke node_akhir)

3. Cobalah kode berikut untuk menemukan jalur terpendek<br>


```python
def find_shortest_path(graph, node_awal, node_akhir, path=[]):
    path = path + [node_awal]
    if node_awal == node_akhir:
        print("===== Jalur yang ditemukan:", path)
        return path
    if node_awal not in graph:
        return None
    shortest = []

    for node in graph[node_awal]:
        print("Path saat ini:", path)
        print("Node tujuan berikutnya adalah", node, "dari node sebelumnya yaitu", node_awal)

        if node not in path:
            newpaths = find_shortest_path(graph, node, node_akhir, path)
            if newpaths:
                if not shortest or len(newpaths) < len(shortest):
                    shortest = newpaths
    return shortest

find_shortest_path(graph, "1","5")
```

**Penjelasan Program**<br>
program  diatas adalah fungsi untuk menemukan jalur ter efektif/terpendek yang mungkin dari `node_awal` ke `node_akhir`<br>
fungsi `find_shortest_path` memiliki 4 parapeter:
- `graph` = ini untuk menyimpan struktur graph
- `node_awal` = untuk menyimpan nilai node awal yg ingin dicari
- `node_akhir` = untuk menyimpan nilai node akhir yang ingin ditemukan jalurnya
- `path` = ini isinya list kosong yg digunakan untuk menyimpan jalur saat kita melakukan pencarian pada program nantinya

isi program ini sama persis dengan fungsi `find_all_path` namun ada perbedaan sedikit di bagian setelah kode `if node not in path:`<br>
setelah memanggil kembali fungsi `find_shortest_path(graph, node, node_akhir, path)` yang disimpan dalam variabel newpaths, maka dilanjutkan:
- `if newpaths:` = cek apakah variabel newpaths itu ada nilainya
  - `if not shortest or len(newpaths) < len(shortest):` = ini untuk cek jika shortest sudah ada isinya atau panjang nilai newpath lebih kecil daripada panjang nilai shortest maka:
  - nilai shortest diganti dengan nilai newpaths
- `return shortest` = menampilkan nilai shortest yang akan menampilkan jalur terpendek yang ditemukan

5. Hitung total jarak yang harus dilalui dari node 1 ke node 5. Tentukan jalur terpendeknya!

```python
def find_distance(graph, node_awal, node_akhir, path=[], distance=0):
    path = path + [node_awal]
    if node_awal == node_akhir:
        return distance
    if node_awal not in graph:
        return None
    
    longest = None
    
    for node in graph[node_awal]:
        if node not in path:
            new_distance = distance + graph[node_awal][node]
            new_path = find_distance(graph, node, node_akhir, path, new_distance)
            if new_path is not None:
                if longest is None or new_path < longest:
                    longest = new_path
    return longest

find_distance(graphh, "1", "5")
```

**Penjelasan Program**<br>
program diatas dibuat untuk menemukan jarak terpendek antara node_awal ke node_akhir yang tiap nodenya memiliki nilai<br>
fungsi ini bernama `find_distance` dengan 5 parameter yaitu:
- `graph` = menyimpan nilai dictionary graphh yang berisi graf dengan node yg memiliki nilai pada tiap edge nya
- `node_awal` = simpul awal
- `node_akhir` = simpul akhir (tujuan)
- `path` = list kosong untuk menyimpan jalur
- `distance` = nilai awal jumlah jarak yang ditempuh, nilainya mula mula 

penjelasan isi fungsi per baris:<br>
ini program bagian pertama mirip dengan fungsi sebelumnya<br>
- `path = path + [node_awal]` = menambahkan nilai node_awal ke path
- `if node_awal == node_akhir:` = jika node_awal samadengan node_akhir maka mengembalikan nilai jaraknya 0
- `if node_awal not in graph:` = jika node awal gaada di graph, maka akan mengembalikan nilai None(jarak tidak ditemukan)
- `longest` = None = inisiasi variavel untuk menyimpan jarak terpendek
- `for node in graph[node_awal]:` = untuk tiap nilai (node) yg ada di `graph[node_awal]`
- `if node not in path:` = jika node tidak ditemukan di path maka 
  - `new_distance = distance + graph[node_awal][node]` = inisiasi variabel baru dengan nilai menambah node_awal dan node yg dikunjungi dengan nilai distance
  - `new_path = find_distance(graph, node, node_akhir, path, new_distance)` = memanggil ulang fungsi dan disimpan ke variabel baru
  - `if new_path is not None:` = jika di variabel new_path ada isinya maka
    - `if longest is None or new_path < longest:` = cek jika variabel longest tidak ada isinya atau nilai new_path lebih kecil dari nilai longest maka akan merubah nilai `longest` menjadi sama dengan `new_path`
- `return longest` = fungsi mengembalikan nilai longest yang berisi panjang jarak terpendek yang bisa ditempuh
