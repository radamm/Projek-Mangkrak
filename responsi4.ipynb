```python
import pandas as pd    
import numpy as np      
import sklearn          
from sklearn.model_selection import train_test_split      
from sklearn.linear_model import LinearRegression           
import matplotlib.pyplot as plt                            
```

```python
df = pd.read_csv("F:/UNAIR/ALGORITMA PEMROGRAMAN/responsilast.csv")        # membaca file format csv sesuai direktori di laptop
df
```

```python
mapping1 = {
    'Terlambat' : 0 ,
    'Tidak terlambat' : 1,
}

mapping2 = {
    'Tidak' : 0 ,
    'Datang' : 1,
}

mapping3 = {
    'Tidak' : 0 ,
    'Ya' : 1,
}

mapping4 = {
    'Belum' : 0 ,
    'Sudah' : 1,
}

mapping5 = {
    'Tidak' : 0 ,
    'Ya' : 1,
}

df['Bangun'] = df['Bangun'].map(mapping1)
df['Perkuliahan'] = df['Perkuliahan'].map(mapping2)
df['Hujan'] = df['Hujan'].map(mapping3)
df['Makan'] = df['Makan'].map(mapping4)
df['Mandi'] = df['Mandi'].map(mapping5)
```

```python
x = df.drop('Perkuliahan', axis=1)
y = df['Perkuliahan']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.3, random_state=100)
```

```python
x_train = np.array(x_train).reshape(-1,4)
x_test = np.array(x_test).reshape(-1,1)
regresilinear = LinearRegression()
regresilinear.fit(x_train,y_train)
```
