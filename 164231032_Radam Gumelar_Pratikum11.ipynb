1. Buatlah sebuah program pembalik kata. Langkah pertama, tiap katakter pada String input diekstrak dan dimasukkan ke dalam stack. Kemudian tiap karakter tersebut dikeluarkan dan ditampilkan sebagai output. Karena memiliki sifat LIFO, maka keluaran stack adalah karakter-karakter dengan urutan yang berkebalikan dengan input. Program ini memiliki 3
module yaitu: <br>
    - Stack untuk menyimpan setiap karakter input pada stack array yang berisi constructor dan function operasi stack
    - Pembalik untuk membaca setiap karakter input, menyimpan karakter dengan memanggil function push() pada class “stack” (point a), dan membalik input dengan memanggil function pop() pada class “stack” (point a). Tiap karakter keluaran dari stack tersebut disimpan/ditambahkan (append) pada String output sebagai keluaran yang akan ditampilkan
    - PembalikKata ini berisi function main untuk deklarasi dan inisialisasi input, memanggil file Pembalik untuk membalik input dan mendapatkan output kata yang telah dibalik, serta menampilkan output pada console

```python
class Stack:
    def __init__(self):
        self.stack    = []  # list kosong
        self.kata     = input("Masukkan kata")  # inputan
        self.reversed = ""  # string kosong

    def push(self):
        if self.kata:
            for i in self.kata:     # untuk tiap karakter di inputan
                self.stack.append(i)    # memasukkan tiap karakter ke dalam list
        else:
            print("Tidak ada kata yang diinputkan!")

    def pop(self):
        while self.stack:
            self.reversed += self.stack.pop()   # mengambil tiap karakter di self.stack kemudian ditambahkan ke self.reversed

    def reversed(self):
        reversed_string = self.reversed     # mengambil nilai di self.reversed
        string          = self.kata     # mengambil nilai di self.stack
        if reversed_string:
            print(f"Kata '{string}' berhasil dibalik menjadi '{reversed_string}'")  # menampilkan output
        else:
            print("Stack kosong!")

#pemanggilan fungsi
a = Stack()
a.push()
a.pop()
a.reversed()
```

- `self.stack` = list kosong
- `self.kata` = inputan
- `self.reversed` = string kosong untuk menyimpan hasil kata yang dibalik
<br>
pada class tersebut memiliki beberapa fungsi yang digunakan untuk membalikkan kata yang digunakan dengan runtut yaitu:
1. fungsi `push(self)` untuk memisah tiap huruf pada inputan `self.kata` kemudian ditambahkan ke `self.stack`
2. fungsi `pop(self)` untuk mengambil nilai/karakter/indeks mulai dari yang paling akhir di list `self.stack` kemudian dimasukkan ke `self.reversed`
3. fungsi `reversed(self)` untuk menampilkan output kata yang berhasil dibalik

2. Buatlah program antrian dengan mengimplementasikan Queue Stack. Adapun aturannya sebagai berikut:
    - penambahan objek pada daftar antrian. Lakukan beberapa kali hingga antrian penuh (maksimum 5). Ketika objek ditambahkan pada antrian yang penuh maka program akan menampilkan keterangan antrian penuh.
    - Menampilkan isi antrian
    - satu persatu objek keluar antrian hingga antrian kosong<br>
<br>


```python
daftar = []
elemen = input("Masukan data antrian: ").split(",")
antrian = int(input("Jumlah antrian: "))

def elemen_masuk():
    while True:  
        i = 0
        for a in range(len(elemen)):
            if len(daftar) < antrian:
                    daftar.append(elemen[i])
                    print(f"{elemen[i]} masuk antrian")
            else:
                print(f"Maaf {elemen[i]}, antrian sudah penuh")
            i+=1
        break

def elemen_keluar():
     while True:
        a = len(daftar)
        while a > 0:
            keluar = daftar.pop(0)
            print(f"{keluar} keluar antrian")
            print(f",".join(daftar+['Kosong']*(antrian-len(daftar))))
            a-=1
        print("Antrian Kosong")
        print(f"Jumlah antrian: {a} orang")
        break

print("> Inputan data antrian")
elemen_masuk()

print("> Satu persatu keluar antrian")
elemen_keluar()
```

3. Buatlah program penelusuran tree dengan menggunakan metode preorder, inorder dan postorder traversal, kemudian tentukan **leaf**, dan **root** dari tree berikut <br>


```python
# Class describing a node of tree
class Node:
	def __init__(self, v):
		self.left = None
		self.right = None
		self.mid  = None
		self.data = v

# Inorder Traversal
def Inorder(node):
    if node is None:
        return
    Inorder(node.left)
    print(node.data,end=" ")
    Inorder(node.mid)
    Inorder(node.right)

# Preorder Traversal
def PreOrder(node):
    if node is None:
        return
    print(node.data, end = " ")
    PreOrder(node.left)
    PreOrder(node.mid)
    PreOrder(node.right)

# Postorder Traversal
def PostOrder(node):
    if node is None:
        return
    PostOrder(node.left)
    PostOrder(node.mid)
    PostOrder(node.right)
    print(node.data, end = " ")

def main(inputan,root):
    if inputan == "postorder":
        return PostOrder(root)
    elif inputan == "preorder":
        return PreOrder(root)
    elif inputan == "inorder":
        return Inorder(root)
```

```python
#untuk membuat tree
root = Node("Z")
root.left = Node("A")
root.left.left = Node("X")
root.left.left.left = Node("B")
root.left.left.mid = Node("N")
root.left.left.right = Node("W")
root.left.left.right.left = Node("Q")
root.mid = Node("G")
root.right = Node("R")
root.right.right = Node("F")
root.right.right.left = Node("S")
root.right.right.left.left = Node("D")
root.right.right.left.right = Node("K")
root.right.right.right = Node("J")
root.right.right.right.right = Node("P")
```

```python
# fungsi untuk pemanggilan
perintah = input("Masukkan pilihan (Postorder/Preorder,Inorder): ").lower()
print(f"{perintah} Traversal: ", end = "")
main(perintah,root)
```
