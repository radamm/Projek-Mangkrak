```python
def kedua(lis):
    if len(lis) <2:
        return "nilai dalam variabel lis kurang dari 2"
    return lis[-2]
kedua(["pentol","nasgor","migor","bubur"])
```

- program dimulai dengan membuat fungsi dengan nama `kedua` dengan 1 parameter dengan nama lis
  - mula mula akan dicek terlebih dahulu jumlah dari nilai yang diinputkan dengan menggunakan `len(lis)`, jika jumlah nilai kurang dari sama dengan 2 maka akan menampilkan tulisan 'nilai dalam variabel lis kurang dari 2'
  - di dalam fungsi, kode `lis[-2]` bermakna mengambil nilai ke 2 dari belakang dari value lis (bisa berupa string ataupun list, namun dalam kasus kali ini inputan yang saya buat adalah list)

b. Fungsi untuk mengkuadratkan setiap elemen pada list yang berisi angka.

```python
def kuadrat(angka):
    return [x **2 for x in angka]
kuadrat([2,4])
```

- program dimulai dengan membuat fungsi dengan nama `kuadrat` dengan 1 parameter dengan nama angka
  - di dalam fungsi, kode `x **2 for x in angka` bermakna mengambil semua nilai dari list angka, tiap nilai tersebut akan dipangkatkan dengan 2
    - penjelasannya:
    - `x**2` berarti memangkatkan nilai dengan 2
    - `for x in angka` berarti mengambil setiap nilai di list angka

sebagai berikut:
<ol>
<li>Fungsi kalkulator() berisi program dengan meminta pengguna untuk memilih operasi matematika yang diinginkan (+, -, /, *)</li>
<li>Tiap operasi matematika memiliki fungsi yang menerima 2 input angka dan mengembalikan hasil operasinya.</li>
<li>Fungsi kalkulator akan menampilkan hasil setelah angka diinputkan oleh pengguna dan memilih operasi matematika yang diinginkan. Jika tidak memenuhi persyaratan yang dimaksud, tampilkan pesan error dan meminta pengguna untuk mengulang proses tersebut.</li>
</ol>

```python
def tambah(a,b):
    return a+b

def kurang(a,b):
    return a-b

def kali(a,b):
    return a*b

def bagi(a,b):
    if b != 0:
        return a/b
    else:
        print("Tidak bisa dibagi dengan nol")
        kalkulator()

def kalkulator():
    while True:
        try:
            a = int(input("Masukkan angka pertama: "))
            b = int(input("Masukkan angka kedua: "))
            try:
                operasi = input("Masukkan operasi (+,-,/,*): ")
                if operasi == "+":
                    return tambah(a,b)
                elif operasi == "-":
                    return kurang(a,b)
                elif operasi == "*":
                    return kali(a,b)
                elif operasi == "/":
                    if b == 0:
                        raise ValueError("Tidak dapat dibagi dengan nol")
                    return bagi(a,b)
            except ValueError:
                print("Tidak ada pilihan operasi")
        except ValueError:
            print("Inputan Bukan angka")
kalkulator()
```

- buat 4 fungsi untuk tiap operasi
  - fungsi `tambah` untuk menjumlahkan variabel a dan b
  - fungsi `kurang` untuk mengurangkan variabel a dan b
  - fungsi `kali` untuk mengalikan variabel a dan b
  - fungsi `bagi` untuk menbagi variabel a dengan b

- program utama ada pada fungsi `kalkulator`
  - saat fungsi dijalankan, user akan diminta menginputkan 2 angka, jika yang diinputkan bukan angka maka akan diminta input ulang dan ada peringatan "inputan bukan angka"
  - kemudian setelah itu, user diminta menginputkan operasi yang diinginkan. jika salah input maka akan diminta menginputkan ulang dan ada peringatan "tidak ada pilihan operasi"
  - jika sudah sesuai maka program akan memanggil fungsi operasi sesuai dengan apa yang diinputkan tadi
